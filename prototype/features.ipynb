```python
import pyrem as pr
import numpy as np
import pylab as pl
import pandas as pd
```

```python
seq = np.genfromtxt("/home/quentin/Desktop/rem_chunking/test_eeg_no_header.txt")
to_remove =10 * int(seq.size / 100.0)
seq = seq[to_remove: -to_remove]
# seq = np.sin(np.linspace(0, 200, 100000)) *10
# seq = seq + np.sin(np.linspace(0, 20, 100000)) * 2
# seq = seq + np.random.normal(0,10,100000)
sig = pr.Signal(seq, 200.0)
```

```python
sig = sig.resample(256)
print sig.duration
sig[0:300000]
```

```python
ff = pr.features.FeatureFactory()
dfs = [ff(t,s) for t, s in sig.embed_seq(sig.sampling_freq * 60, sig.sampling_freq * 20 )]
features = pd.concat(dfs)
features.shape
```

```python
pl.plot(features["wavelets.db4_mean_cA_5"]/features["wavelets.db4_mean_cD_5"],'-')
pl.ylim(0,2)
#pl.plot(features["wavelets.db4_median_cD_4"] , '-')
#pl.plot(features["power_sd"],'-')
```

```python
f = np.array(features)
```

```python
from sklearn import cluster

```

```python
km = cluster.KMeans(4)
fit = km.fit(f)
```

```python
fit
```

```python

p = km.predict(f)
color_map = ["k","b","r","g"]
pl.figure()
for i, (c, x, y) in enumerate(zip(p, features["power_median"], features["power_median"])):
    pl.plot(i, x,'.',color=color_map[c])

```
